### Polymorphism
allows objects of different classes to be treated as objects of a common superclass

Provide way to perform a single action in different forms

achieved through method overriding and interfaces

### Polymorphism with Abstract base class (ABC)
Abstract base classes are used to define common methods for a group of related objects. They can enforce tthe derived classes to implement particular methods, promoting consistency across different implementations



```python
### Runtime Polymorphism
class Animal():
    def speak(self):
        print("Animal speak")
class Dog(Animal):
    def speak(self):
        print("Dog speak")
class Cat(Animal):
    def speak(self):
        print("meow")

d = Dog()
d.speak()

### Compile time polymorphism
def add(a,s,d):
    return a+s+d
def add(a,s):
    return a+s
print(add(1,2))

### Abstract base class
from abc import ABC, abstractclassmethod

## Abstract class
class Vehicle(ABC):
    @abstractclassmethod
    def start_eng(self):
        pass
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"
    
```

### Encapsulation and Abstraction
involves bundling data and methods that operate on the data within single unit, while abstraction involves hiding complex implementation details and exposing only the necessary features

### Encapsulation
wraps data and methods together as a single unit. Restricts data access to some of the objects components prevent data misuse
Using getter and setter
public - accessible anywhere
private - use __ (double underscore to define private variables)
protected - use _ (single underscore to define protected)


```python
class Person:
    def __init__(self, name, age):
        self.__name = name   ## it is a private variable
        self.__age = age
    
    def getter(self):
        return self.__name


def get_name(person):
    return person.name

person = Person("Krish",33)
person.getter()
```

### Magic methods 
in Python also know as dinder methods(double underscore methods) are special methods that start and end with double underscores. These methods unable you to define the behavior of the objects for built in operations such as arithmetic , comparison etc4

dir(person) - gives the list of methods

```python
'''
__init__ - Initializes a new instance of class
__str__ - Returns a string representation of object
__repr__ - returns an official string rep of an object
__len__ - Returns the length of an object
__getitem__: gets an item from a container
__setitem__: sets an item in a container
'''
class Person():
    def __init__(self, age):
        self.age = age
    def __str__(self):
        return self.age
person = Person("10")
print(person)
# dir(person)
```

### Operator overloading 
using different magic methods like add, mul, sub to perform operations

```python
class Vector():
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):
        return f"{self.x}     {self.y}"
    
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)

v1 = Vector(1,2)
v2 = Vector(3,4)
print(v1+v2)

```
